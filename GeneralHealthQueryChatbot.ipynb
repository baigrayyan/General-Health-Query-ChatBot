### General Health Query Chatbot

This section will guide you through creating a simple chatbot using OpenAI's GPT-3.5 model. The chatbot will act as a helpful medical assistant, providing general health information while explicitly avoiding actual medical advice.

**To get started, you'll need an OpenAI API key.**
1.  Go to [OpenAI API Keys](https://platform.openai.com/account/api-keys).
2.  Create a new secret key.
3.  In Colab, go to the left panel, click the '🔑' icon (Secrets), and add a new secret named `OPENAI_API_KEY` with your OpenAI API key as the value.

```python
import openai
from google.colab import userdata

# Retrieve the API key securely from Colab's user data
OPENAI_API_KEY = userdata.get('OPENAI_API_KEY')
openai.api_key = OPENAI_API_KEY

def get_health_response(query):
    # Prompt engineering to define the chatbot's persona and safety guidelines
    prompt = f"""You are a friendly and helpful medical assistant. Your goal is to provide general, factual information about health-related questions in a clear and easy-to-understand manner. You must NEVER give medical advice, diagnoses, or treatment recommendations. Always advise the user to consult a qualified healthcare professional for any medical concerns.

Here is the user's question:
'{query}'

Based on the above, provide a general information response. Remember to be helpful, clear, and always include a disclaimer about seeking professional medical advice."""

    try:
        response = openai.chat.completions.create(
            model="gpt-3.5-turbo",
            messages=[
                {"role": "system", "content": prompt},
                {"role": "user", "content": query}
            ],
            max_tokens=200,
            temperature=0.7
        )
        return response.choices[0].message.content
    except Exception as e:
        return f"An error occurred: {e}. Please ensure your API key is correctly set and you have an active internet connection."

# Example Usage
print("Welcome to the General Health Info Chatbot!")
print("Ask me anything general about health. Type 'quit' to exit.")

while True:
    user_query = input("\nYour question: ")
    if user_query.lower() == 'quit':
        print("Goodbye!")
        break

    bot_response = get_health_response(user_query)
    print(f"Bot: {bot_response}")
```
